* Count Plot

* Box Plot: Phương pháp để mô tả bằng đồ thị các nhóm dữ liệu số thông qua các phần tử của chúng
* Join Plot
* Pair Plot

### Distribution Plot
* Univariate distribution of observations

```python
tips_df = sns.load_dataset('tips')
```

```python
tips_df.head()
```

```python
sns.histplot(data=tips_df['total_bill']);
```

```python
tips_df['total_bill'].value_counts().sort_values(ascending=False)
```

### KDE Plot
* A Kernel Density Estimate (KDE) plot is a method for visualising the distribution of observation in a dataset, analagous to a histogram
* KDE represents the data using a continuous probability density curse in one or more dimensions

```python
sns.kdeplot(data=tips_df['total_bill'])
```

### Displot 
* Informative distributional summaries
* Statistical analysis require knowledge about the distribution of variables in your dataset.
* The seaborn function `displot()` support serveral approaches to visualising distributions. These include classic techniques like histograms and computationally-intensive approaches like kenel density estimation.

```python
sns.displot(data=tips_df, x='total_bill', col='time', kde=True);
```

## Categorical Plots

#### Bar Plots

```python
tips_df.head()
```

```python
sns.barplot(data=tips_df, x='sex', y='tip', estimator=np.mean)
```

#### Count Plots

```python
tips_df['sex'].value_counts()
```

```python
sns.countplot(data=tips_df, x='sex')
```

#### Box Plot
Box Plot is a good indication of how the values in the data are spread out ("minimum", first quartile (Q1), median, third quartile (Q3), and "maximum")
* Identify outliers
* Check if data is symmetrical
* Determine how tightly data is grouped


```python
sns.boxplot(data=tips_df, x='day', y='total_bill', hue='sex', palette='Blues')
plt.legend(loc=0)
```

### Facet Grid
* Facet Grid can be drawn with up to three dimensions: `row`, `col` and `hue`.
* The main approach for visualising data on this grid is with the `FacetGrid.map()` method.

```python
tips_df.head()
```

```python
tips_fg = sns.FacetGrid(data=tips_df, row='smoker', col='time') # Create a class instance of Facet Grid class
tips_fg.map(sns.scatterplot, 'total_bill', 'tip')
```

```python
# Create a keywords dictionary (kws)
kws=dict(s=100, edgecolor='r', alpha=.5)

new_fg=sns.FacetGrid(data=tips_df, col='sex', 
                     hue='smoker',
                     col_order=['Female', 'Male'],
                     palette='Set2',
                     height=4,
                     aspect=1.4)
new_fg.map(sns.scatterplot, 'total_bill', 'tip', **kws)
new_fg.add_legend();
```

### Join Plot 
* Join distribution between two variables along with each variable's marginal distribution

```python
penguins_df = sns.load_dataset('penguins')
```

```python
penguins_df.head()
```

```python
sns.jointplot(data=penguins_df, x='flipper_length_mm', y='bill_length_mm', hue='species')
```

### Pair Plots

```python
sns.pairplot(data=penguins_df, hue='species')
```

### Heatmaps
Describing density or intensity of variables, visualize patterns, variance, and even anomalies

```python
flights_df = sns.load_dataset('flights')
```

```python
flights_df.head()
```

```python
flights = pd.pivot_table(flights_df, index='month', columns='year', values='passengers')
```

```python
flights
```

```python
sns.heatmap(data=flights, cmap='Blues', linecolor='white', linewidths=1)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
cereal_df = pd.read_csv('./data/cereal.csv') 
```

```python
cereal_df.head()
```

```python
# irrelevant fields
fields=['shelf', 'weight', 'cups', 'rating']
cereal_df_new = cereal_df.drop(fields, axis=1)
```

```python
cereal_df_new.head()
```

```python
# corr()
# Mối tương quan của các dữ liệu trong dataframe
cereal_corr = cereal_df_new.corr() # Get correlation data
cereal_corr
```

```python
# ones_like can build a metrix of booleans (True, False) with the same shape as our data
ones_corr = np.ones_like(cereal_corr, dtype=bool)
ones_corr
```

```python
ones_corr.shape, cereal_corr.shape
```

```python
# np's triu: return only upper triangle matrix
mask = np.triu(ones_corr)
mask
```

```python
sns.heatmap(data=cereal_corr, mask=mask);
```

```python
# When removing the upper tri-angle, here are stil two enpty cells in our matrix (calories and vitamins)
adjusted_mask = mask[1:, :-1]
adjusted_mask
```

```python
adjusted_cereal_corr = cereal_corr.iloc[1: , :-1]
```
